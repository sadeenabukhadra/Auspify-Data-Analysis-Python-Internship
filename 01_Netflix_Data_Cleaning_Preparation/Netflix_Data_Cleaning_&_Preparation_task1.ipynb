```python
## 4. Duplicate Records Analysis
"""
Duplicate records were checked to identify repeated rows that may affect analysis accuracy.

"""
```

```python
# Check complete duplicate rows
df.duplicated().sum()
```

```python
### Observation
"""
No complete duplicate records were found in the dataset.
Therefore, no rows were removed.
"""
```

```python
# Check duplicated titles
df[df.duplicated('title', keep=False)].sort_values('title')
```

```python
### Observation
"""
Some titles appear more than once; however, these records have different show IDs.

Example:
- Same title with different show_id values.

These are not duplicate records because each row represents a separate entry.
Therefore, they were kept.
"""
```

```python
## 5. Data Standardization
"""
Text columns were checked and cleaned to remove formatting inconsistencies.

"""
```

```python
# Remove extra spaces from text columns
text_columns = df.select_dtypes(include = 'object').columns
for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
df.columns
```

```python
### Country Column Standardization
"""
The country column was checked for:
- Extra spaces.
- Different capitalization formats.

The values were already standardized, so only formatting consistency was ensured.

"""
```

```python
df['country'].value_counts()
```

```python
df['country'] = df['country'].str.strip()
df['country'] = df['country'].str.title()
```

```python
print("Check the first 10 unique country values to verify consistency after standardization\n")
print(df['country'].unique()[:10])
# Check the frequency of each country category
print("\n  Check distribution")
df['country'].value_counts()
```

```python
### Rating Column Standardization
"""
The rating column was standardized by:
- Removing unnecessary spaces.
- Ensuring uppercase formatting.

The values were already consistent.
"""
```

```python
df['rating'].value_counts()
```

```python
df['rating'] = df['rating'].str.strip()
df['rating'] = df['rating'].str.upper()

print("Check the first 10 unique rating values to verify consistency after standardization")
print(df['rating'].unique()[:10])
```

```python
print("\n Check distribution")
df['rating'].value_counts()
```

```python
### Type Column Standardization
"""
The type column was checked for inconsistent category names.

The values were already standardized as:
- Movie
- TV Show
"""
```

```python
df['type'].value_counts()
```

```python
from os import replace
df['type'] = df['type'].str.strip()
df['type'] = df['type'].replace(
    {
        'Tv Show':'TV Show',
        'tv Show':'TV Show',
        'movie':'Movie'

    }
)
print("Check the first 10 unique type values to verify consistency after standardization")
df['type'].unique()
```

```python
print("\n Check distribution")
df['type'].value_counts()
```

```python
## 6. Final Dataset Validation
"""
After cleaning and standardization, the dataset was checked again to ensure:
- No missing values.
- No duplicate records.
- Correct data formats.
"""
df.info()
df.describe(include='all')
```
